```python
import cv2
import os
import subprocess
```

# Download YouTube Video

```python
video_url = "https://www.youtube.com/watch?v=rn4cjaqxO3c"
output_video = "front_side.mp4"
!yt-dlp -f best -o {output_video} {video_url}
```

```python
# change front_side with yt video name
output_dir = "front_side"
os.makedirs(output_dir,exist_ok=True)
```

```python
cap = cv2.VideoCapture(output_video)
fps = int(cap.get(cv2.CAP_PROP_FPS))
print(fps)
```

```python
frame_interval = 30
```

# Extract Frames from the Video

```python
frame_count = 0
saved_count = 0

while cap.isOpened():
  ret, frame = cap.read()

  if not ret:
    break
  if frame_count % frame_interval == 0:
    img_name = f"{output_dir}/frame_{saved_count}_{output_dir}.jpg"
    cv2.imwrite(img_name,frame)
    saved_count += 1
  frame_count += 1
```

```python
cap.release()
print(f"Dataset created with {saved_count} images in '{output_dir}'")
```
